```python
#since columns names hasnt done in a correct way we are reading the dataset again
col_names=['target','id','date','flag','user','text']
twitter_data=pd.read_csv('training.1600000.processed.noemoticon.csv',names=col_names,encoding='ISO-8859-1')
```

```python
#converting the targets from 4 to 1 , to have a better preprocessing
twitter_data.replace({'target':{4:1}},inplace=True)
```

```python
twitter_data['target'].value_counts()
```

```python
#0 means negative tweet and 1 means positive tweet
#stemming (cutting down of word to a root word)
port_stem=PorterStemmer()
```

```python
def stemming(content):
    stemmed_content=re.sub('[^a-zA-Z]',' ',content)
    stemmed_content=stemmed_content.lower()
    stemmed_content=stemmed_content.split()
    stemmed_content=[port_stem.stem(word) for word in stemmed_content if not word in stopwords.words('english')]
    stemmed_content=''.join(stemmed_content)
    return stemmed_content
```

```python
twitter_data['stemmed_content']=twitter_data['text'].apply(stemming)
```

```python
twitter_data.head()
```

```python
print(twitter_data['stemmed_content'])
```

```python
print(twitter_data['target'])
```

```python
#sepearting the data and label
x=twitter_data['stemmed_content'].values
y=twitter_data['target'].values
```

```python
print(x)
print("\n")
print(y)
```

```python
#splitting the data as training and test data
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,stratify=y,random_state=2)
```

```python
print(x.shape,x_test.shape,x_train.shape)
```

```python
#ml doesnt understand languages since we need to convert this into numerical values
vectorizer=TfidfVectorizer()
x_train=vectorizer.fit_transform(x_train)
x_test=vectorizer.transform(x_test)
print(x_train,x_test)
```

```python
#training the ml model
#logistic regression
model=LogisticRegression(max_iter=1000)
model.fit(x_train,y_train)
```

```python
#model evaluation 
#accuracy score
x_train_pred=model.predict(x_train)
training_data_acc=accuracy_score(y_train,x_train_pred)
```

```python
print('Accuracy score ',training_data_acc)
```

```python
x_test_pred=model.predict(x_test)
testing_data_acc=accuracy_score(y_test,x_test_pred)
```

```python
print('Accuracy score ',testing_data_acc)
```
